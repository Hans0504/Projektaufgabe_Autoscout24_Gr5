####  1. Autoscout24 csv Datei einlesen

```python
import pandas as pd
import locale as lo

df = pd.read_csv("Datensatz_Aufgabe_1_Autoscout24_Germany.csv", dtype={"price":int})

df.head()
```

#### 2. Überschriften verändern (deutsch)!

```python
df.columns = ['Kilometerstand', 'Marke', 'Modell', 'Kraftstoff', 'Getriebe', 'Zustandsart', 'Preis', 'PS', 'Baujahr']
```

#### 4. Spaltenwerte anzeigen lassen (Anzahl)!

```python
werte_kraftstoff = df['Kraftstoff'].value_counts()
werte_zustandsart = df['Zustandsart'].value_counts()
werte_getriebe = df['Getriebe'].value_counts()
```

#### 5. Spalte Kraftstoff/Getriebe/Zustandsart auf deutsch übersetzten!

```python
df['Kraftstoff'] = df['Kraftstoff'].replace({
    'Diesel': 'Diesel',
    'Gasoline': 'Benzin',
    'Electric/Gasoline': 'Hybrid',
    'Electric/Diesel': 'Hybrid',
    'Electric': 'Elektro',
    'Others': 'sonstiges',
    '-/- (Fuel)': 'sonstiges',
    'LPG': 'sonstiges',
    'CNG': 'sonstiges',
    'Ethanol': 'sonstiges',
    'Hydrogen': 'sonstiges'  
})

df['Getriebe'] = df['Getriebe'].replace({
    'Manual': 'Manuell',
    'Automatic': 'Automatik',
    'Semi-automatic': 'Halb-Automatik'
})

df['Zustandsart'] = df['Zustandsart'].replace({
    'Used': 'Gebraucht',
    'Pre-registered': 'Erstzulassung',
    "Employee's car": 'Firmenwagen',
    'New': 'Neu',
    'Demonstration':'Vorführwagen'
})
```

#### 6. Überprüfung ob alle Werte richtig übersetzt sind ! (Filtern)

```python
filter_df = df[df['Marke'] == 'Audi']
```

#### 7. Duplikate bereinigen!

```python
df_cleaned = df.drop_duplicates(subset=['Kilometerstand',
                                        'Marke',
                                        'Modell',
                                        'Kraftstoff',
                                        'Getriebe',
                                        'Zustandsart',
                                        'Preis',
                                        'PS',
                                        'Baujahr'
                                        ])
```

#### 8. Überprüfung der Bereinigung

```python
filter_cl_df = df_cleaned[df_cleaned['Marke'] == 'Fiat']
filter_cl_df
```

```python
df_cleaned.to_csv("df_autoscout-cl.csv", sep=",", index=False)
```

#### 10. csv-Dateien zusammenfügen!

```python
# Einladen des Exceldatensatzes vom KBA

# Lesen des Excel-Datensatzes in einen Pandas-Dataframe
df_kba = pd.read_excel("KBA_2024_04_31_V2.xlsx")

# Anzeigen der ersten 5 Zeilen des Dataframes
df_kba
```

```python
# Löschen der überflüssigen Spalten (nur die ersten 2 werden benötigt)

# Löschen der Spalten "Unnamed: 2" bis zum Ende
df_kba = df_kba.iloc[:, :2]
```

```python
# Schleife, um NaN-Werte in der ""Segment"" Spalte zu ersetzen
for i in range(len(df_kba)):
    if pd.isna(df_kba.at[i, "Segment"]):
        if df_kba.at[i-1, "Segment"] == "MINI":
            df_kba.at[i, "Segment"] = "MINI"
        elif df_kba.at[i-1, "Segment"] == "KLEINWAGEN":
            df_kba.at[i, "Segment"] = "KLEINWAGEN"
        elif df_kba.at[i-1, "Segment"] == "KOMPAKTKLASSE":
            df_kba.at[i, "Segment"] = "KOMPAKTKLASSE"
        elif df_kba.at[i-1, "Segment"] == "MITTELKLASSE":
            df_kba.at[i, "Segment"] = "MITTELKLASSE"
        elif df_kba.at[i-1, "Segment"] == "OBERE MITTELKLASSE":
            df_kba.at[i, "Segment"] = "OBERE MITTELKLASSE"
        elif df_kba.at[i-1, "Segment"] == "OBERKLASSE":
            df_kba.at[i, "Segment"] = "OBERKLASSE"
        elif df_kba.at[i-1, "Segment"] == "SUVs":
            df_kba.at[i, "Segment"] = "SUVs"
        elif df_kba.at[i-1, "Segment"] == "GELÄNDEWAGEN":
            df_kba.at[i, "Segment"] = "GELÄNDEWAGEN"
        elif df_kba.at[i-1, "Segment"] == "SPORTWAGEN":
            df_kba.at[i, "Segment"] = "SPORTWAGEN"
        elif df_kba.at[i-1, "Segment"] == "MINI-VANS":
            df_kba.at[i, "Segment"] = "MINI-VANS"
        elif df_kba.at[i-1, "Segment"] == "GROSSRAUM-VANS":
            df_kba.at[i, "Segment"] = "GROSSRAUM-VANS"
        elif df_kba.at[i-1, "Segment"] == "UTILITIES":
            df_kba.at[i, "Segment"] = "UTILITIES"
        elif df_kba.at[i-1, "Segment"] == "WOHNMOBILE":
            df_kba.at[i, "Segment"] = "WOHNMOBILE"
        else:
            df_kba.at[i, "Segment"] = df_kba.at[i-1, "Segment"]
```

```python
# Entfernen aller Zeilen mit NaN-Werten
df_kba = df_kba.dropna()

#Zeilen mit "Sonstige" entfernen
df_kba = df_kba[df_kba["Modellreihe"].str.contains("SONSTIGE") == False]


```

```python
# Teilen der "Modellreihe" Spalte an dem ersten Leerzeichen

# Split "Modellreihe" in "Marke" und "Typ", ab dem 1. Leerzeichen und nur 1x
df_kba[["Marke", "Typ"]] = df_kba["Modellreihe"].str.split(r"\s+", n=1, expand=True)
```

```python
#Speichern als CSV
df_kba.to_csv("KBA_Typenliste.csv", sep=",", index=False)
```

```python
# Die fehlenden Werte in Segment werden von df_kba in df_combined übernommen, sofern diese vorhanden sind
df_combined["Segment"] = df_combined["Segment"].fillna(df_kba["Segment"])

# # Entferne die überschüssigen Spalten aus df_kba (Modellreihe, Modellreihe und Typ)
df_combined = df_combined.drop(["Modellreihe", "Marke_y", "Typ"], axis=1)
```
